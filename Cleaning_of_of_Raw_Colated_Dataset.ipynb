```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option("max_rows", 100)
```

```python
Phenotype_Soil_Weather_Data_RAW = pd.read_csv("/content/drive/MyDrive/University of reading/MSC FINAL project/Data/Data Files/Finalised data/Phenotype_Soil_Weather_Data_RAW_EXP.csv")
```

```python
# this as the soil and weather data was key for the use in the prediction algoritum all rows that did not include soil data where dropped as there is no set location for them.
Phenotype_Soil_Weather_Data_Cleaned1 = Phenotype_Soil_Weather_Data_RAW.dropna(subset=["Temperature [C]","Location","Grain Yield (bu/A)"])
```

```python
Phenotype_Soil_Weather_Data_Cleaned1.count()
```

```python
Phenotype_Soil_Weather_Data_Cleaned2 = Phenotype_Soil_Weather_Data_Cleaned1.dropna(thresh=len(Phenotype_Soil_Weather_Data_Cleaned1) - 1500, axis=1)
```

```python
Phenotype_Soil_Weather_Data_Cleaned2.count()
```

```python
Phenotype_Soil_Weather_Data_Cleaned3 = Phenotype_Soil_Weather_Data_Cleaned2.drop('Unnamed: 0', 1)
```

```python
Phenotype_Soil_Weather_Data_Cleaned3 = Phenotype_Soil_Weather_Data_Cleaned3.reset_index(drop=True)
```

```python
Phenotype_Soil_Weather_Data_Cleaned3
```
