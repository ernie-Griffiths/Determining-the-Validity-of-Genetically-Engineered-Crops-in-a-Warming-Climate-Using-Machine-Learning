```python
# these columns are not uniform in type.
#Plot Discarded [enter 'yes' or blank]
#Comments
#Filler [enter 'filler' or blank]
#Possible subs
#Confirmed subs
#Date Plot Planted [MM/DD/YY]
#Date Plot Harvested [MM/DD/YY]
#Anthesis [MM/DD/YY]
#Silking [MM/DD/YY]
thing = type(Phenotype_data["Date Plot Planted [MM/DD/YY]"][0])
print(thing)
print("<this is the original>")
for i in Phenotype_data["Date Plot Planted [MM/DD/YY]"]:
  if thing != type(i):
    print(i)
    print(type(i))
```

```python
Phenotype_data[['Plant Month',"Plant Day","Plant year"]] = Phenotype_data['Date Plot Planted [MM/DD/YY]'].str.split("/", 2, expand=True)
Phenotype_data[['Harvest Month',"Harvest Day","Harvest year"]] = Phenotype_data['Date Plot Harvested [MM/DD/YY]'].str.split("/", 2, expand=True)
```

```python
#'Anthesis [MM/DD/YY]','Silking [MM/DD/YY]', Removed from the dropping as these values Corispond to phenotypical attributes.
Phenotype_data = Phenotype_data.drop(['Year','State','City','Date Plot Planted [MM/DD/YY]','Date Plot Harvested [MM/DD/YY]','Plant Day','Plant year','Harvest Day','Harvest year'], 1)
Phenotype_data['Plant Month'] = pd.to_numeric(Phenotype_data['Plant Month'])
Phenotype_data['Harvest Month'] = pd.to_numeric(Phenotype_data['Harvest Month'])
```

```python
Soil_data = Soil_data.drop(['Grower','Date Received', 'Date Reported','Comments'], 1)
```

```python
Phenotype_Soil_data = pd.merge(Phenotype_data, Soil_data, how='left', left_on = 'Field-Location', right_on = 'Location')
```
