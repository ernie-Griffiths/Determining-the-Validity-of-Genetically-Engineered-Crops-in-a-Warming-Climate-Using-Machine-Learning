With certain colums of the weather dataset required modification , changing to from strings to floats, this would allow for averages to be taken that could then be applied to the phenotypical dataset.

Changing the Wind hust from an object containing strings and floats into just floats.

```python
Weather_data.loc[Weather_data['Wind Gust [m/s]'] == "#REF!", "Wind Gust [m/s]"] = 0
Weather_data['Wind Gust [m/s]'] = Weather_data['Wind Gust [m/s]'].astype(np.float64)
```

#Breaking the weather dataset up, into chunks.

```python
Location_array = Weather_data["Field Location"].unique()
```

```python
Location_array
```

```python
LocationDict = dict(tuple(Weather_data.groupby('Field Location')))
```

```python
LocationDict[Location_array[1]]
```

Further breaking the dataframe into chunks based on the months,

```python
#i= 5
#TempMonthDataset = LocationDict[Location_array[i]].groupby('Month')[["Field Location","Month","Temperature [C]", "Dew Point [C]", "Relative Humidity [%]","Solar Radiation [W/m2]",	"Rainfall [mm]",	"Wind Speed [m/s]",	"Wind Direction [degrees]",	"Wind Gust [m/s]",	"Soil Temperature [C]",	"Soil Moisture [%VWC]",	"Soil EC [mS/cm]",	"UV Light [uM/m2s]",	"PAR [uM/m2s]",	"CO2 [ppm]"]].mean()
#TempMonthDataset["Field Location"] = Location_array[i]
#TempMonthDataset
```

```python
MonthedData = pd.DataFrame()
for i in range(len(Location_array)):
  TempMonthDataset = LocationDict[Location_array[i]].groupby('Month')[["Field Location","Month","Temperature [C]", "Dew Point [C]", "Relative Humidity [%]","Solar Radiation [W/m2]",	"Rainfall [mm]",	"Wind Speed [m/s]",	"Wind Direction [degrees]",	"Wind Gust [m/s]",	"Soil Temperature [C]",	"Soil Moisture [%VWC]",	"Soil EC [mS/cm]",	"UV Light [uM/m2s]",	"PAR [uM/m2s]",	"CO2 [ppm]"]].mean()
  TempMonthDataset["Field Location"] = Location_array[i]
 # print(i)
  MonthedData = pd.concat([MonthedData, TempMonthDataset], axis=0, ignore_index= True)
```
